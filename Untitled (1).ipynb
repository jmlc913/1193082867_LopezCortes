```python
5. Escribir un programa que reciba una cadena de texto y reporte:
#   a) cuántas letras vocales en mayúscula se entraron;
#   b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
#   c) cuántos dígitos se entraron,
#   d) cuántos espacios se entraron;
#   e) cuántas palabras reservadas se entraron.
#   Utilizar para ello los métodos de cadenas especializados como .isupper(), 
#    .isdigit(), etc. y la palabra reservada in.

import keyword
numeroTildes=0
numeroEspacios=0
numeroReservadas=0
numeroNumeros=0
numeroMayusculas=0
codigos=[205,218,225,193,233,201,243,211,250,237]
reservadas=keyword.kwlist

frase=input("Escriba ")

for n in frase:
    
    o=ord(n)
    if o in codigos:
        
        numeroTildes+=1
    
    if n.isspace():
        
        numeroEspacios+=1
    
    if n.isdigit():
        
        numeroNumeros+=1
    
    if n.isupper():
        
        numeroMayusculas+=1

for i in range(0,len(reservadas)):
    
    if reservadas[i] in frase:
        
        numeroReservadas+=1
        
print("mayusculas=",numeroMayusculas,",tildes=",numeroTildes,"espacios=",numeroEspacios)
print("palabras reservadas=",numeroReservadas,",numeros=",numeroNumeros)
```
